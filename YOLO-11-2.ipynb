# Testing-video-3

```python
# Load YOLOv8 model
model = YOLO('/kaggle/working/runs/detect/train18/weights/best.pt')  # You can also use other models like yolov8s.pt or yolov8m.pt

# Directory with frames and a new directory for predictions
frames_dir = '/kaggle/input/pexels-testing-data/manybirds-frames'
predicted_frames_dir = '/kaggle/working/manybirds_-output-11-new-4'
os.makedirs(predicted_frames_dir, exist_ok=True)
```

```python
for frame_file in sorted(os.listdir(frames_dir)):
    frame_path = os.path.join(frames_dir, frame_file)
    frame = cv2.imread(frame_path)

    # Predict objects in the frame
    results = model(frame, conf=0.3)

    # Extract detections and count objects
    object_count = len(results[0].boxes)  # Total detections in frame

    # Manually draw bounding boxes without class names
    for box in results[0].boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # Extract box coordinates
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Draw bounding box

    # Add object count overlay
    cv2.putText(frame, f'Objects detected: {object_count}', (10, 30), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

    # Save the annotated frame
    predicted_frame_path = os.path.join(predicted_frames_dir, frame_file)
    cv2.imwrite(predicted_frame_path, frame)
```

```python
# Define output video path with .mp4 extension
output_video_path = '/kaggle/working/manybirds_video-11-new-4.mp4'

# Set frames per second and frame size
fps = 30  # Adjust as needed
frame_size = (frame.shape[1], frame.shape[0])  # Width, height of frames

# Create a VideoWriter object for MP4 (H.264 codec)
video_writer = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, frame_size)

# Write each annotated frame to the video
for predicted_frame_file in sorted(os.listdir(predicted_frames_dir)):
    predicted_frame_path = os.path.join(predicted_frames_dir, predicted_frame_file)
    frame = cv2.imread(predicted_frame_path)
    video_writer.write(frame)

# Release the VideoWriter object
video_writer.release()
print("Annotated MP4 video saved at:", output_video_path)
```

```python
import matplotlib.pyplot as plt
import os

# Set the path to your YOLO output directory
output_dir = '/kaggle/working/runs/detect/train15'  # Adjust this path if needed

# Function to display an image with a specific figure size
def display_image(image_name):
    image_path = os.path.join(output_dir, image_name)
    if os.path.exists(image_path):
        plt.figure(figsize=(14, 7))  # Set the figure size
        img = plt.imread(image_path)   # Load the image
        plt.imshow(img)                # Display the image
        plt.axis('off')                # Hide axes
        plt.title(image_name)          # Title the image
        plt.show()                     # Show the plot
    else:
        print(f"Image {image_path} does not exist.")

# Display the necessary curve images
display_image('R_curve.png')   # Replace with the actual image file names
display_image('PR_curve.png')
display_image('F1_curve.png')
display_image('P_curve.png')
```
